```python
import numpy as np
from matplotlib import pyplot as plt

#---------------------------------------------------
N = 10000 # Number of samples
a = 0.1 # Lower extreme
b = 10 # Higher extreme
nbins = int(N / (b - a) / 2)

# Uniform distribution between a and b. The corresponding distribution is 1/(b-a)
x = np.random.uniform(a, b, N)

# Probability Density Function (PDF) of x
def pdf_x(low, high):
    return 1 / (high - low)

# Transformation of the initial data
# The corresponding distribution is (ln(10)*10^y)*1/(b-a)
y = np.log10(x)

# PDF of y
def pdf_y(y, low, high): #PDF of y
    return np.log(10) * 10 ** y * pdf_x(low, high)

#----------------------------------------------------------------
# Plotting the PDF of x and y
fig, ax = plt.subplots(1, 2, figsize=(16, 9))

# Plot of PDF of x
ax[0].hist(x, nbins, density=True)
ax[0].set_title("PDF of x")
ax[0].set_xlabel("x")

# Plot of PDF of y
ax[1].hist(y, nbins, density=True)
ax[1].set_title("PDF of y")
ax[1].set_xlabel("y")

#-------------------------------------------------------
# Plotting the theoretical PDF of y
sample = np.log10(np.linspace(a, b, N)) # Sample for the theoretical PDF
ax[1].plot(sample, pdf_y(sample, a, b), label="Theoretical PDF")
ax[1].legend()

plt.show()

#Printing Mean and Median of the two distributions

print("Log_10(Mean of x) = " + str(np.log10(np.mean(x))))
print("Mean of y = " + str(np.mean(y)))
print("\nLog_10(Median of x) = " + str(np.log10(np.median(x))))
print("Median of y = " + str(np.median(y)))

```
